```python
import pandas as pd
```

```python
canada_covid = pd.read_csv('clean_data/cleaned-canada-covid.csv', parse_dates=['date'])
canada_covid.info()
canada_covid.sort_values('date')
```

```python
canada_vaccine = pd.read_csv('clean_data/cleaned-canada-vaccination.csv', parse_dates=['week_end'])
canada_vaccine.info()
canada_vaccine.sort_values('week_end')
```

```python
canada_covid_1 = canada_covid[canada_covid['date'] <= pd.to_datetime('20201212')]
canada_covid_1.sort_values('date')
canada_covid_2 = canada_covid[canada_covid['date'] > pd.to_datetime('20201212')]
canada_covid_1.sort_values('date')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```
